```python
from datasets import load_dataset
dataset = load_dataset("Muennighoff/flores200", 'all')
print(dataset)
print(dataset['dev'][0])
```

```python
# Two datasets are available, 'dev' and 'devtest'.
# Each data sample have the sentences in multiple languages.
for data in dataset['dev']:
  # Here I print out one example in the 'dev' set.
  print(data)
  for itm in data:
    print(itm, ':', data[itm])

  # Select the sentecen pairs based on your interest of languages.
  # The code for each language is available on this page:
  # https://github.com/facebookresearch/flores/blob/main/flores200/README.md
  # Here I select English and French. and print out the corresponding sentecens.
  print('sentence in english: \n', data['sentence_eng_Latn']) #English
  print('sentence in french: \n', data['sentence_fra_Latn']) #French
  break
```
